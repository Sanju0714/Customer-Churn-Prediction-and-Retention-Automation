```python
import pandas as pd

# Load dataset
df = pd.read_csv("swiggy_customer_data.csv")

print("===== DATA VALIDATION REPORT =====\n")

# Rule 1: Customer ID should be unique
print("Customer ID Unique:",
      "PASS" if df["customer_id"].is_unique else "FAIL")

# Rule 2: Age should be between 18 and 60
print("Age Validation:",
      "PASS" if df["age"].between(18, 60).all() else "FAIL")

# Rule 3: Orders should be >= 0
print("Orders Validation:",
      "PASS" if (df["orders_last_30_days"] >= 0).all() else "FAIL")

# Rule 4: Avg Order Value > 0
print("Avg Order Value Validation:",
      "PASS" if (df["avg_order_value"] > 0).all() else "FAIL")

# Rule 5: Churn should be 0 or 1
print("Churn Validation:",
      "PASS" if df["churn"].isin([0,1]).all() else "FAIL")

# Rule 6: Valid Cities
valid_cities = ["Bangalore","Mumbai","Chennai","Hyderabad"]

print("City Validation:",
      "PASS" if df["city"].isin(valid_cities).all() else "FAIL")

# Missing Values Check
print("\nMissing Values:")
print(df.isnull().sum())

# Duplicate Check
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
import pandas as pd

df = pd.read_csv("swiggy_customer_data.csv")

errors = []

if df["customer_id"].isnull().sum() > 0:
    errors.append("customer_id column has null values")

if df["age"].isnull().sum() > 0:
    errors.append("age column has null values")

if df["city"].isnull().sum() > 0:
    errors.append("city column has null values")

if df["orders_last_30_days"].isnull().sum() > 0:
    errors.append("orders_last_30_days column has null values")

if df["avg_order_value"].isnull().sum() > 0:
    errors.append("avg_order_value column has null values")

if len(errors) == 0:
    print("Dataset Passed Validation")
else:
    print("Errors Found:")
    for error in errors:
        print(error)
```
